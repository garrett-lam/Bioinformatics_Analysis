# Hypothesis


Our group has two main hypotheses:
- **Null hypothesis (H0):** There is no significant relationship between the number of pregnancies, cholesterol levels, blood pressure, BMI, age, and skin thickness and the likelihood of developing diabetes in the Pima Indian female population. 
- **Alternative hypothesis (H1):** There is at least one health factor (number of pregnancies, cholesterol levels, blood pressure, BMI, age, skin thickness) that has a significant relationship with the likelihood of developing diabetes in the Pima Indian female population.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; We believe BMI, blood pressure, cholesterol levels will have a significant relationship with oneâ€™s likelihood of developing diabetes in the Pima Indian female population, whereas number of pregnancies, skin thickness, and age will have an insignificant relationship with one's likelihood of developing diabetes in the Pima Indian female population. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; There is evidence that BMI, blood pressure, and cholesterol levels are all risk factors for the development of diabetes. BMI is a measure of body fat based on a person's height and weight. Studies have consistently shown that people with higher BMIs are at an increased risk of developing diabetes, as excess body fat can impair the body's ability to use insulin effectively. High blood pressure is also a known risk factor for diabetes, it can damage blood vessels and lead to insulin resistance, which is a key factor in the development of diabetes. Cholesterol levels have also been linked to an increased risk of diabetes. High levels of cholesterol can contribute to the buildup of plaque in the arteries, which can impair blood flow and increase the risk of developing insulin resistance. Therefore, it is reasonable to believe that there is a significant relationship between BMI, blood pressure, cholesterol levels, and the likelihood of developing diabetes [^diabetesjournals].

### K-Nearest Neighbors (KNN) 

Finally, we decided to build a K-nearest neighbor (KNN) machine learning model. The KNN algorithm is a non-parametric, supervised learning algorithm often used for classification of data. Our group decided to use KNN as it is a simple machine learning algorithm that classifies a point by comparing it to itâ€™s k nearest neighbors. By using this algorithm, we can better predict the chances of a patient having diabetes by comparing the new patientâ€™s data to the existing database. From the EDA conducted above, we determined that the variables: Age, SkinThickness, and BloodPressure are not statistically significant. Therefore, we decided to drop these variable from our dataset before we run KNN. 

The first step of KNN is to normalize the data. This is done using the Standard Scalar method which normalizes data based on its z-score.


```python
# This line creates an instance of the StandardScaler class, which will be used to scale the data.
scalar = StandardScaler()

# This line applies the fit_transform method of the StandardScaler object to the data in the df dataframe, excluding the columns 
# "Age", "SkinThickness", "BloodPressure", and "Diabetes". The resulting scaled and transformed data is stored in a new dataframe 
# called transformed, with columns named "Pregnancies", "Glucose", "Insulin", "BMI", and "DiabetesPedigreeFunction".
transformed =  pd.DataFrame(scalar.fit_transform(df.drop(["Age","SkinThickness","BloodPressure","Diabetes"],axis = 1),), 
        columns=['Pregnancies', 'Glucose', 'Insulin','BMI', 'DiabetesPedigreeFunction'])

# This line creates a new variable called dependent that contains the Diabetes column from the original df dataframe. 
# This column will be used as the dependent variable in a machine learning model.
dependent = df.Diabetes
```

```python
# Print the first couple observations of the transformed dataframe
transformed.head()
```

We are going to split the data randomly into training and testing data using 70/30 split. The train_test_split() function ensures that training and testing data are randomly selected.

```python
# This line applies the train_test_split function to the scaled and transformed data (transformed) and the dependent 
# variable (dependent). The resulting split data is stored in four separate variables: transformed_train, transformed_test, 
# dependent_train, and dependent_test.
# transformed_train: This variable contains the training dataset for the scaled and transformed data.
# transformed_test: This variable contains the testing dataset for the scaled and transformed data.
# dependent_train: This variable contains the training dataset for the dependent variable.
# dependent_test: This variable contains the testing dataset for the dependent variable.
# The test_size parameter is set to 0.30, which means that 30% of the data will be used for testing and 
# 70% will be used for training. The random_state parameter is set to 42, which ensures that the same random splits 
# are generated every time the code is run. The stratify parameter is set to dependent, which ensures that the training and 
# testing datasets have the same proportion of values for the Diabetes column as the original dataset. 
# This is particularly useful when the dependent variable is imbalanced.
transformed_train,transformed_test,dependent_train,dependent_test = train_test_split(transformed,dependent,test_size=0.30,random_state=42, stratify=dependent)
```

```python
# Containers to hold testing and training data
test = []
train = []

for i in range(1,12):
    # This line creates an instance of the KNeighborsClassifier class with i neighbors.
    knn = KNeighborsClassifier(i)
    # This line fits the KNN model to the training dataset, using the fit method of the KNeighborsClassifier object.
    knn.fit(transformed_train,dependent_train)
    
    # These lines use the score method of the KNeighborsClassifier object to calculate the accuracy score of the model 
    # on the training and testing datasets, respectively. The resulting scores are then appended to the train and test lists.
    # At the end of this loop, the train and test lists will contain the accuracy scores for the KNN model on the training and testing datasets, 
    # respectively, for values of i ranging from 1 to 11. These scores can be used to evaluate the performance of the model 
    # and select the optimal value of i.
    train.append(knn.score(transformed_train,dependent_train))
    test.append(knn.score(transformed_test,dependent_test))
```

We will use a lineplot to determine the best k value to use for our model

```python
# This line creates a new figure object with a width of 10 inches and a height of 5 inches, using the figure method of the 
# pyplot module from the matplotlib library.
#plt.figure(figsize=(10,5))

# This line creates a line plot using the lineplot method of the Seaborn (sns) library. The range(1,12) function is used to 
# specify the x-axis values (i.e., the number of neighbors used in the KNN model), and the train list is used to specify the 
# y-axis values (i.e., the accuracy scores on the training dataset). The label parameter is used to set the label for the line 
# to "Train", which will be used in the plot legend.
fig, ax = plt.subplots()
sns.lineplot(x=range(1, 12), y=train, label='Train', ax=ax)
sns.lineplot(x=range(1, 12), y=test, label='Test', ax=ax)
```

From the graph we can see that the max testing score occurs when k = 11. We will use this k to to make the model.

```python
# Setup a knn classifier with k neighbors
knn = KNeighborsClassifier(11)

# This line fits the KNN model to the training dataset using the fit method of the KNeighborsClassifier object. 
# The transformed_train and dependent_train variables contain the feature and target variables, respectively, 
# for the training dataset.
knn.fit(transformed_train,dependent_train)

# This line evaluates the performance of the KNN model on the testing dataset using the score method of the 
# KNeighborsClassifier object. The transformed_test and dependent_test variables contain the feature and target variables, 
# respectively, for the testing dataset. The resulting score variable contains the accuracy score of the KNN model on the 
# testing dataset. By calling the fit method on the KNeighborsClassifier object with the training dataset, we are training 
# the model to learn the relationships between the feature variables and the target variable. Once the model is trained, 
# we can evaluate its performance on the testing dataset by calling the score method with the testing dataset. 
# The resulting accuracy score provides a measure of how well the model generalizes to new, unseen data.
score = knn.score(transformed_test,dependent_test)

print(f"The best test score produced by our model is {score}, where k = 11")
```

To evaluate our KNN Model, we will be using a confusion matrix to calculate recall, precision, and accurarcy of our model.

```python
# Let us get the predictions using the classifier we had fit above
# This line uses the predict method of the KNeighborsClassifier object to make predictions on the testing dataset. 
# The transformed_test variable contains the feature variables for the testing dataset. The resulting prediction 
# variable contains the predicted target variable values.
prediction = knn.predict(transformed_test)

# This line uses the confusion_matrix function from the sklearn.metrics module to generate a confusion matrix for 
# the predictions. The dependent_test variable contains the true target variable values for the testing dataset, 
# and the prediction variable contains the predicted target variable values. The resulting confusion matrix is a 
# table that summarizes the number of true positives, true negatives, false positives, and false negatives for the 
# model's predictions.
confusion_matrix(dependent_test,prediction)

# This line creates a cross-tabulation table using the crosstab function from the pandas library. The dependent_test 
# variable contains the true target variable values for the testing dataset, and the prediction variable contains the 
# predicted target variable values. The resulting table shows the number of actual and predicted values for each class 
# in the target variable, as well as the total counts (margins) for each row and column. The rownames and colnames 
# parameters are used to specify the names of the rows and columns in the table, respectively.
pd.crosstab(dependent_test, prediction, rownames=['Predicted'], colnames=['Actual'], margins=True)
```

```python
# Recall is a measure of how many of the actual positive results were correctly identified by the model. 
# It is calculated as the number of true positives (TP) divided by the sum of true positives and false negatives (FN). 
# A high recall score indicates that the model produces few false negatives.
recall = 104/(104+30)

# Precision is a measure of how many of the predicted positive results were actually positive. 
# It is calculated as the number of true positives (TP) divided by the sum of true positives and false positives (FP). 
# A high precision score indicates that the model produces few false positives.
precision = 104/(104+4)

# Accuracy is a measure of how often the model makes correct predictions. It is calculated as the number of true positives 
# (TP) plus true negatives (TN) divided by the total number of predictions. A high accuracy score indicates that the model 
# produces few false positives and false negatives.
accuracy = (104 + 24)/(104+4+24+30)

print(f"Recall: {recall}")
print(f"Precision: {precision}")
print(f"Accuracy: {accuracy}")
```
